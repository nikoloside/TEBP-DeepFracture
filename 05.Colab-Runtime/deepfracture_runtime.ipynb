# 💥 DeepFracture Runtime — Colab demo

Neural brittle-fracture prediction inside a rigid-body runtime, end to end:

1. **Collision detection** — PyBullet simulates a projectile hitting the target and captures the strongest contact.
2. **Collision embedding** — the impact becomes the paper's 7-D condition *(position, direction, impulse)* in the target's local frame.
3. **Neural prediction** — the per-shape VQ-VAE ([nikoloside/deepfracture](https://huggingface.co/nikoloside/deepfracture)) decodes a geometrically-segmented SDF.
4. **Segmentation** — pure-Python watershed/flooding (scikit-image) extracts one mesh per fragment.
5. **Back to the runtime** — the intact body is swapped for the fragments, which keep simulating.

Papers: [DeepFracture (CGF 2025)](https://nikoloside.graphics/deepfracture/) · [Far-From-Boundary Fields (SMI 2026)](https://nikoloside.graphics/far-from-boundary-fields/)

> Runtime → **Run all** takes about 2 minutes on a free CPU runtime. No GPU needed.
